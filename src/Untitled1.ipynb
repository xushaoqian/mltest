```python
import numpy as np

X = np.array([[0.1,0.1,0.1],[0.9,0.9,0.1],[0.2,0.2,0.1],[0.8,0.8,0.1],[0.1,0.05,0.1],[0.9,0.95,0.1]])
y = np.array([[0,1,0,1,0,1]]).T 
syn1 = 2*np.random.random((3,1)) - 1
print(syn1)
```

```python
for j in range(1000):
    l1 = 1/(1+np.exp(-(np.dot(X,syn1)))) 
    l1_delta = (y - l1)*(l1*(1-l1))  
    syn1 += X.T.dot(l1_delta)
    print(syn1)
```
